### Code to impute attributes- these are the preferences that the respondent has to particular questions

```python
# Dependencies and Setup

import numpy as np
import pandas as pd

# File to Load - Limit to waves 1-5 and 10-21 for 100 point allocation
file_to_load = "Resources/Speed_Dating_Data_wavelimit.csv"

# Read Speed Dating File and store into Pandas data frame
date_df = pd.read_csv(file_to_load)

#show first 5 records of dataframe
date_df.head()
```

```python
#don't overwrite the original dataframe
date1_df = date_df
```

### Impute the mean by each attribute for each missing attribute

```python
#Fill missing values with the mean 
for column in subset:
    date1_df[column].replace(np.NaN, date1_df[column].mean(), inplace=True)
```
